```python
import os
import cv2
import numpy as np
import pandas as pd

from PIL import Image

def letterbox_image(image, size):
    '''resize image with unchanged aspect ratio using padding'''
    iw, ih = image.size
    w, h = size
    scale = min(w/iw, h/ih)
    nw = int(iw*scale)
    nh = int(ih*scale)

    image = image.resize((nw,nh), Image.BICUBIC)
    new_image = Image.new('RGB', size, (128,128,128))
    new_image.paste(image, ((w-nw)//2, (h-nh)//2))
    return new_image

def _preprocess_data(image_path, csv_path):
    csv = pd.read_csv(csv_path)
    labels = True if csv.shape[1] == 5 else False
    for image in train['image_id']:
        path = image_path + image + '.jpg'
        im = Image.open(path)
        im = np.array(letterbox_image(im, (600,600)))
        im = np.reshape(im,[1]+list(im.shape))
        if 'X' not in locals():
            X = im
        else:
            X = np.concatenate((X,im))

    X = X/255
    if labels:
        y = csv.loc[:, 'healthy':].values
        return X,y
    else:
        return X
    
image_path = 'C:/Users/emage/OneDrive/Desktop/apple_k/data/images/'
csv_path = 'C:/Users/emage/OneDrive/Desktop/apple_k/data/train.csv'
x,y = _preprocess_data(image_path, csv_path)
print(x.shape)
```
